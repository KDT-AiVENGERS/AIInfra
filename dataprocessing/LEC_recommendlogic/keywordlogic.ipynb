```python
import pandas as pd

def Rec_Lec(Rec_JDs : pd.DataFrame, keywords : list, Lec_udemy : pd.DataFrame):
    Rec_JDs = Rec_JDs.reset_index(drop = True)
    result = []
    for idx1 in range(len(Rec_JDs)):
        string = Rec_JDs['자격요건'][idx1] + Rec_JDs['우대조건'][idx1] + Rec_JDs['주요업무'][idx1]
        string = string.lower().replace(' ', '')
    # 난이도 필터링

        k_list = []
        for keyword in keywords:
            if keyword in string:
                k_list.append(keyword)

        n = len(Lec_udemy)
        
        Rec_dict = dict()
        for idx2 in range(n):

            Udemy_string = Lec_udemy.loc[idx2, '강의소개'].lower().replace(' ', '')
            count = 0
            for key in k_list:
                count += Udemy_string.count(key)
                Rec_dict[idx2]=count
        
        Rec_dict = dict(sorted(Rec_dict.items(), key = lambda x: x[1]))
        l = list(Rec_dict.keys())[:3] # 10개 추천으로 작성

        print(idx1, '번 공고에 대한 추천 강의')
        print(l)
        print('-'*40)
        display(Lec_udemy.iloc[l, :])

        result.append((idx1, Lec_udemy.iloc[l, :]))

    return result
```

```python
udemy_skills = pd.read_csv('./dataset/udemy_skills.csv')
jd_skills = pd.read_csv('./dataset/job_skills.csv')

kw = pd.merge(udemy_skills, jd_skills, on = '단어명', how = 'inner')['단어명'].values
kw
```
